### Getting AAPL stock Options related data including Implied Volatilities

```python
import yfinance as yf
from datetime import datetime
import pandas as pd

# Define the ticker symbol
mb_tickerSymbol = 'AAPL'

# Get data on this ticker
mb_tickerData = yf.Ticker(mb_tickerSymbol)

# Get the option expirations
mb_options_expirations = mb_tickerData.options

# Prepare a list to hold all options data
mb_options_list = []

# Loop through each options expiration date
for expiration in mb_options_expirations:
    # Get options data for each expiration date
    mb_options_data = mb_tickerData.option_chain(expiration)
    
    # Add option type and expiration date to the data
    mb_options_data.calls['OptionType'] = 'Call'
    mb_options_data.calls['expirationDate'] = expiration
    mb_options_data.puts['OptionType'] = 'Put'
    mb_options_data.puts['expirationDate'] = expiration
    
    # Append the data to the options_list
    mb_options_list.append(mb_options_data.calls)
    mb_options_list.append(mb_options_data.puts)

# Concatenate all options data into a single DataFrame
mb_all_options_data05192023 = pd.concat(mb_options_list)

# Reset the index and drop the old one
mb_all_options_data05192023.reset_index(drop=True, inplace=True)

# Convert the expirationDate column to datetime
mb_all_options_data05192023['expirationDate'] = pd.to_datetime(mb_all_options_data05192023['expirationDate'])

# Calculate the time to expiration in days
mb_all_options_data05192023['time_to_expiration'] = (mb_all_options_data05192023['expirationDate'] - datetime.now()).dt.days

# Display the DataFrame
print(mb_all_options_data05192023)
```

```python
# Copy Sources Data DataFrame
mb_prices = mb_src_prices.copy()

# Drop the unwanted columns from the DataFrame
mb_prices = mb_prices.drop(['Open', 'High', 'Low', 'Volume', 'Dividends', 'Stock Splits'], axis=1)

# Format the index to display only the date in 'yyyy-mm-dd' format
mb_prices.index = pd.to_datetime(mb_prices.index).strftime('%Y-%m-%d')

# Rename the 'Close' column to 'AAPL'
mb_prices = mb_prices.rename(columns={'Close': 'AAPL'})

# Print the updated DataFrame with formatted index
print(mb_prices)
```

#### mb_prices enriched

```python
import numpy as np

# Calculating the log returns
mb_prices['Log Returns'] = np.log(mb_prices['AAPL'] / mb_prices['AAPL'].shift(1))

# Number of trading days in a year
trading_days = 252

# Calculating the 1-year rolling volatility
mb_prices['1y Historical Volatility'] = mb_prices['Log Returns'].rolling(window=trading_days).std() * np.sqrt(trading_days)

# Drop the 'Log Returns' column as we don't need it anymore
mb_prices = mb_prices.drop(columns=['Log Returns'])

print(mb_prices)
```

### Implied Volatilities

```python
# Take a Copy of the Options DataFrame
df = mb_options.copy()

# Filter for call options and create a pivot table
call_options = df[df['OptionType'] == 'Call']
call_iv_matrix = call_options.pivot(index='time_to_expiration', columns='strike', values='impliedVolatility')

# Filter for put options and create a pivot table
put_options = df[df['OptionType'] == 'Put']
put_iv_matrix = put_options.pivot(index='time_to_expiration', columns='strike', values='impliedVolatility')

# Fill NaN values with zero (optional)
call_iv_matrix = call_iv_matrix.fillna(0)
put_iv_matrix = put_iv_matrix.fillna(0)

# Sort the index and columns (optional)
call_iv_matrix.sort_index(axis=0, inplace=True)
call_iv_matrix.sort_index(axis=1, inplace=True)

put_iv_matrix.sort_index(axis=0, inplace=True)
put_iv_matrix.sort_index(axis=1, inplace=True)
```

```python
# Select first 5 and last 5 columns of Put implied volatility matrix
put_selected = pd.concat([put_iv_matrix.iloc[:, :5], put_iv_matrix.iloc[:, -5:]], axis=1)
print("\nSelected columns from Put implied volatility matrix:")
put_selected
```

#### mb_iv

```python
'''
Let's now combine the 2 matrices in to one mb_iv
The rule here is to take the average of the 2 when the vol is not zero or not n/a
'''

# First replace zeros with NaN in both matrices
call_iv_matrix.replace(0, np.nan, inplace=True)
put_iv_matrix.replace(0, np.nan, inplace=True)

# Define a custom function to calculate average if neither is zero or NaN
def custom_combine(ser1, ser2):
    ser1_valid = ser1.notna()
    ser2_valid = ser2.notna()
    both_valid = ser1_valid & ser2_valid
    return pd.Series(np.where(both_valid, (ser1 + ser2)/2, np.where(ser1_valid, ser1, ser2)), index=ser1.index)

# Apply the custom function column-wise
mb_iv = call_iv_matrix.combine(put_iv_matrix, custom_combine)

print("Combined implied volatility matrix:")
print(mb_iv)
```

```python
# Select first 5 and last 5 columns of implied volatility matrix
iv_selected = pd.concat([mb_iv.iloc[:, :5], mb_iv.iloc[:, -5:]], axis=1)
print("\nSelected columns from Put implied volatility matrix:")
iv_selected
```

```python
# function to convert tenor to number of days
def tenor_to_days(tenor):
    value = int(tenor[:-1])
    if 'm' in tenor:
        return value * 30  # approximate a month to 30 days
    elif 'y' in tenor:
        return value * 365  # approximate a year to 365 days

# create Tenor column and set as new index
mb_interest_rates['Tenor'] = mb_interest_rates['US_Treasury_Rates'].apply(tenor_to_days)
mb_interest_rates.set_index('Tenor', inplace=True)

# Renaming the Original Pillar Columns
mb_interest_rates.rename(columns={"US_Treasury_Rates": "Pillar"}, inplace=True)

print(mb_interest_rates)
```
